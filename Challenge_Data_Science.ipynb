```python
import pandas as pd
import json
import requests
import numpy as np
```

Acessar Dados - Churn de Clientes

```python
url='https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json'
response = requests.get(url)
dados=response.json()
```

Task 1 - Entender quais informações o conjunto de dados possui


```python
dados_normalizados=pd.json_normalize(dados)
```

```python
dados_normalizados.head()
```

Task 2 - Analisar quais os tipos de dados

```python
dados_normalizados.info()
```

Task 3 - Verificar quais são as inconsistências nos dados

```python
for col in dados_normalizados.columns:
    print(f"Coluna: {col}")
    print(dados_normalizados[col].unique())
    print("-" * 30)
```

```python
dados_normalizados[dados_normalizados['Churn']=='']
```

```python
dados_normalizados[dados_normalizados['customer.tenure']==0]
```

Task 4  - Corrigir as inconsistências nos dados

```python
filter=dados_normalizados['Churn']==''
dados_normalizados.drop(dados_normalizados[filter].index, inplace=True)
```

```python
dados_normalizados[dados_normalizados['Churn']=='']
```

```python
dados_normalizados['account.Charges.Total'] = dados_normalizados['account.Charges.Total'].replace(' ',np.nan).astype(float)
```

```python
dados_normalizados.info()
```

Task 5 - Criar coluna de contas diárias

```python
dados_normalizados['account.Charges.Daily'] = dados_normalizados['account.Charges.Monthly'] / 30
```
